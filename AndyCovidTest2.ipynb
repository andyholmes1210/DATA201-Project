## Trust in Health Industry by Region

Have simplified the function extracting and wrangling the original data and have got it working on all 4 available datasets


```r
# function to wrangle average health trust data by region for different xlsx files so can be plotted against each other
regions = c('Northland', "Auckland", "Waikato", "Bay of Plenty", "Gisborne/Hawke's Bay", 'Taranaki', 'Manawatu-Whanganui', 'Wellington', 
            'Nelson/Tasman/Marlborough/West Coast', 'Nelson/Tasman/ Marlborough/West Coast', 'Canterbury', 'Otago', 'Southland')

extract_trust <- function(wellbeing_period){
    
    #read in wellbeing xlsx, 5th sheet(Regions), skipping first 5 rows due to NA values
    wellbeing_df <- read_xlsx(wellbeing_period, sheet=5, skip=5) 
    
    #selecting columns for regions and their data only, checking that column names are in provided region set
    regionalwb_df <- wellbeing_df %>% select_if(colnames(wellbeing_df) %in% regions) 
    
    #finding row index num for row containing data on average trust (mean row is 5 rows 'lower' than the title row)
    health_trust_row_num <- which(wellbeing_df %>% select(1) == 'Trust held for health system') + 5
    
    #dataframe of just ave rating row, using found row number
    ave_trust_df <- regionalwb_df[health_trust_row_num,] 
    
    #transforming into tidy data with Region col title and regions values in col
    ave_trust_df <- ave_trust_df %>% gather('Regions', Average_rating, 1:12) 
    
    #changing col data type from chr to dbl for plotting functionality
    ave_trust_df <- ave_trust_df %>% mutate(Average_rating = as.double(Average_rating)) 
    
    return(ave_trust_df)
}
```

```r
#loading in wellbeing statistics files
suppressMessages(march21_trust <- extract_trust('wellbeing-statistics-March-2021-quarter.xlsx'))
suppressMessages(dec20_trust <- extract_trust('wellbeing-statistics-december-2020-quarter.xlsx'))
suppressMessages(sept20_trust <- extract_trust('wellbeing-statistics-september-2020-quarter.xlsx'))
suppressMessages(june20_trust <- extract_trust('wellbeing-statistics-june-2020-quarter.xlsx')) #hiding output of default col names
```

```r
#fixing rogue row value in Region in Sept 2020 df

sept20_trust[sept20_trust =='Nelson/Tasman/ Marlborough/West Coast'] <- 'Nelson/Tasman/Marlborough/West Coast'
```

```r
#function to add col of dates for each quarter tibble

add_date_col <- function(trust_quarter, n, value){
    date <- rep(value, n)
    trust_quarter <- trust_quarter %>% add_column(date)
    return(trust_quarter)
}
```

```r
#giving each df a date col

march21_trust <- march21_trust %>% add_date_col(n=nrow(march21_trust), value='March 2021')
dec20_trust <- dec20_trust %>% add_date_col(n=nrow(dec20_trust), value='Dec 2020')
sept20_trust <- sept20_trust %>% add_date_col(n=nrow(sept20_trust), value='Sept 2020')
june20_trust <- june20_trust %>% add_date_col(n=nrow(june20_trust), value='June 2020')
```

### Final data model of regional trust in health industry

```r
#binding all the df's into one

Regionaltrust_df <- bind_rows(june20_trust, sept20_trust, dec20_trust, march21_trust)
Regionaltrust_df %>% head()
```

```r
#plot of the change in average trust of the health system for regions over quarterly periods from june 2020 to march 2021

options(repr.plot.width=14, repr.plot.height=8)

Regionaltrust_plot <- Regionaltrust_df %>% 
    mutate(date= fct_relevel(date, 'June 2020', 'Sept 2020', 'Dec 2020', 'March 2021')) %>% #setting order for x-axis
    ggplot(aes(x=date, y=Average_rating, group=Regions, color = Regions)) + 

    geom_line(lwd=1, position=position_jitter(w=0, h=0.01)) + #jitter lines so double ups are visible
    expand_limits(y = c(5, 9)) + geom_point(color='black') +
    
    ggtitle('Average Trust in Health System by Regions') + #plot tile
    theme(plot.title = element_text(hjust=0, size=22, face='bold'), #theme for labels/title/legend
            legend.key.size = unit(2, 'cm'), 
            legend.key.height = unit(.6, 'cm'),
            legend.key.width = unit(.6, 'cm'), 
            legend.title = element_text(size=20), 
            legend.text = element_text(size=14), 
            axis.text.x = element_text(size=14), 
            axis.text.y = element_text(size=14), 
            axis.title.x = element_text(size=18),
            axis.title.y = element_text(size=18)) +

    xlab('Yearly Quarter') + #x and y labels
    ylab('Average Rating (Out of 10)') 
    
Regionaltrust_plot
```

## Small write up on what can go into the 'deliverable' on this data and its output

We were interested in finding data on wellbeing or attitudes within the population and seeing how and if these had changed over the past 2 years because of Covid. Pertinent data was found on the Stats NZ website, with xlsx files on wellbeing available for the periods of June, September, and December of 2020, and March of 2021.

These files were chosen as they weren't presented in a immediately usable format and would require a certain amount of wrangling and tidying; there is a large amount of empty cells and half-used columns, and so would require some work to get into a usuable and presentable format. At the same time the different files were mostly consistent in their formatting, and so the coding that was created for one file would work on all the others (with some discrepancies to account for).

At the same time, the files also contained data that was interesting and that we thought would be valuable to our 'story'. What was found in the data was the average rating of trust in NZ's health system by region, and so we were interested to see how this had changed, if at all, over the available time frames.

Difficulties arose when initially we attempted to collect the data on different ratings of trust in the health system as opposed to the average trust. This was due to the rows being identified in a non-unique manner. Indexing into the rows themselves wouldn't work due to discrepenancies between files on where these rows actually existed. A workable solution was found by breaking up and merging the identifying columns and finding a specific row identifier, and its following rows. Though given the seemingly arbitrary nature of the different ratings and the difficulty posed in how to present these in an understandable manner, we changed to just collecting the data that was relevant to just the average trust rating.

The techniques used on each file returns a tidy dataframe of just the regions of interest, and their average rating of trust in the health system.
To do this the correct columns need to be extracted from the files, which was acheived using a 'select_if' function, to take only the columns that were titled to the regions of interest. Removal of NA rows first had to be done so that the region values themselves became column names.
The gather function was also used after the desired data was extracted to turn the column regional names into values within their own column, pairing with their extracted average trust value.

Each of these tidy frames was then ammended to have an additional column attached to contain the relevant date of the data, and subsequently these seperate dataframes were combined into one so that all the data can be passed into a graphing function where we can group by the different regions, plotting each of their average trust values and their change over time.

We managed to achieve this for all the relevant datasets available, and thus managed to produce a plot demonstrating how the average trust in NZ's health system has changed over a 12 month time period, and how this has changed for different regions within NZ. What we found was that average trust had decreased from June 2020 to March this year. Presuming future datasets follow a specific format, there should be no issue with incorporating newer datasets with what already exists.
What we have failed to do was combine this with the vaccination rates data due to a lack of overlap of the time periods for the data itself, and so we have failed to see how and if the average trust has increased or continued to decrease as vaccines have been introduced.

## Merging NZ case data with monthly jobseeking data by week

```r
covid_cases <- read_csv('WHO-COVID-19-global-data.csv')
covid_cases %>% head()
```

```r
#removing incorrect date from jobseeker data so correct dates can be taken from cases data

monthly_jobseeking_data_df <- monthly_jobseeking_data_df %>% filter(Date != '2022-09-03') 
```

```r
#getting the cases data that has dates coinciding with jobseeker data

cases_in_weekparameters <- covid_cases %>% filter(Date_reported >= monthly_jobseeking_data_df$Date[1], 
                              Date_reported <= monthly_jobseeking_data_df$Date[length(monthly_jobseeking_data_df$Date)])

#filtering cases data for NZ and only needed cols

nzcases_weekly_df <- cases_in_weekparameters %>% filter(Country == 'New Zealand') %>%
    select(Date_reported, New_cases)
```

```r
#summarizing all the daily case numbers into weekly sums

nzcases_weekly_df <- nzcases_weekly_df %>% 
    mutate(week = cut.Date(Date_reported, breaks = "1 week", labels = FALSE)) %>% 
    arrange(Date_reported) %>%
    group_by(week) %>%
    summarise(WeeklyCases = sum(New_cases), Week = max(Date_reported)) %>%
    select(Week, WeeklyCases)

nzcases_weekly_df %>% head()
```

```r
#adjusting weekly data to match jobseeker data week date

nzcases_weekly_df <-nzcases_weekly_df %>% mutate(Week = Week-2)
```

```r
#summarizing the total jobseekers for each week (ignoring regional differences)

jobseeker_weekly_sum <- monthly_jobseeking_data_df %>% group_by(Date) %>% summarise(Total_jobseeker = sum(Jobseeker_benefit))
```

```r
#fixing last week date as it doesn't add to a whole week

nzcases_weekly_df[length(nzcases_weekly_df$Week),1] = as.Date('2021-09-10')

# changing col name in jobseeker date so can be joined with cases data

colnames(jobseeker_weekly_sum)[which(names(jobseeker_weekly_sum) == "Date")] <- "Week"
```

```r
# fixing col type so can join correctly, then joining jobseeker data with nz cases data on shared col 'Week'

nzcases_weekly_df <- nzcases_weekly_df %>% mutate(Week = as.character(Week))
jobseeker_weeklycases_df <- nzcases_weekly_df %>% inner_join(jobseeker_weekly_sum, by='Week')

#Final relation of joint data

jobseeker_weeklycases_df %>% head()
```

```r
#plotting of weekly total jobseeker benefit data against weekly amount of covid cases

options(repr.plot.width=16, repr.plot.height=8)

jobseeker_weeklycases_plot <- jobseeker_weeklycases_df %>% 
    ggplot(aes(x=Week)) + #setting x axis variable

    geom_point(aes(y=WeeklyCases*500, color='Covid Cases'), shape=19, size=4) + #points for covid cases/scaled to fit to right axis
    geom_point(aes(y=Total_jobseeker, color='Jobseeker Support'), size=4, shape=19) + #points for jobseeker num
    

    theme(axis.text.x = element_text(angle = 75, vjust = 1, hjust = 1)) + #Rotate Date label on x-axis 

    ggtitle("Weekly Amount of People on Jobseeker Support versus Weekly Covid Cases") +

    theme(plot.title = element_text(hjust=0.6, size=24, face='bold'), #theme for labels/title/legend 
            legend.text = element_text(size=14), 
            axis.text.x = element_text(size=14), 
            axis.text.y = element_text(size=14), 
            axis.title.x = element_text(size=18),
            axis.title.y = element_text(size=18)) +

    scale_y_continuous(
    
        #left axis title
        name = "Total Amount on Jobseeker Benefit",
    
        #right axis title and scaling
        sec.axis = sec_axis(~./500, name="Weekly Number of Covid Cases")) + 
    
    theme(legend.title=element_blank()) + #removing legend title
    xlab('Date of Week') #x axis title 

jobseeker_weeklycases_plot
```
